```python
import pandas as pd
import os
import string
```

```python
# Get csv filenames
csvs = [f for f in os.listdir() if f.endswith('.csv')]
```

```python
# Load all csvs into dataframes
dfs = [pd.read_csv(csv, encoding='ISO-8859-1') for csv in csvs]
```

```python
# Per df, list all the columns needed to be extracted
# 1) Death counts, 2) Start year
death_cols = {
    0: ['BatDeath', 'NonStateDeaths'],
    1: ['TotalCombatDeaths'],
    2: ['SideADeaths', 'SideBDeaths'],
    3: ['BatDeath']
}

year_cols = {
    0: ['StartYear1'],
    1: ['StartYear'],
    2: ['StartYear1'],
    3: ['StartYear1']
}
```

```python
# Remove all punctuation from dfs
bad_puncs = [c for c in string.punctuation.replace('.', '')]
remove_these = {}
for c in bad_puncs:
    remove_these[c] = ''
```

```python
# Concat all data into one large df
new_dfs = []
for i, df in enumerate(dfs):
    df = df.replace({',', ''}, regex=True)
    try:
        for col in death_cols[i]:
            df[col] = df[col].str.replace(',', '').astype(float)
        df['total_deaths'] = df[death_cols[i]].sum(axis=1)
        df['start_year'] = df[year_cols[i]]
    except:
        df['total_deaths'] = df[death_cols[i]].sum(axis=1)
        df['start_year'] = df[year_cols[i]]
    new_dfs.append(df[['start_year','total_deaths']])
    
deaths_by_year = pd.concat(new_dfs).groupby('start_year').sum().clip_lower(0)
```

```python
deaths_by_year.plot()
```

```python
deaths_by_year.cumsum().to_csv('../deaths-by-year_cumulative.csv')
```
